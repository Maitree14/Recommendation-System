```python
output = pd.pivot_table(data=data2, 
                        index=['Customer ID'], 
                        columns=['Product Name'], 
                        values='counts',fill_value=0)
output
```

```python
out_list=output.columns
out_list=out_list.tolist()
out_list
```

```python
from scipy.sparse import csr_matrix
df_matrix = csr_matrix(output.values)
from sklearn.neighbors import NearestNeighbors
model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(df_matrix)
```

```python
recommendation = pd.DataFrame(columns = ['ProductId','Recommendation_Percentage','Customer ID'])
main_list = []
for j in range(0,793):
    temp_list=[]
    distances,suggestions= model_knn.kneighbors(output.iloc[j, :].values.reshape(1,-1), n_neighbors =5)
    #print(suggestions.tolist())
    out_l=suggestions.tolist()
    out_l1=out_l.pop(0)
    #print(out_l1)
    out_l1.pop(0)
    #print(out_l1)
    for k1 in out_l1:
        temp_list.append(out_list[k1])
    #print(temp_list)
    main_list.append(temp_list)
print(main_list)
```
